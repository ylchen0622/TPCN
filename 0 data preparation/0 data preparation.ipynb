```python
import pandas as pd
import matplotlib.pyplot as plt
from rdkit import Chem
```

```python
df=pd.read_csv('20230515_All.csv',encoding='GB18030')
```

### Normalize structures and filter out compounds without structures

```python
Standard_Smiles=[]
for i in range(len(df)):
    try:
        canonical_smiles=df['Standard_Smiles'][i]
        canonical_mol=Chem.MolFromSmiles(canonical_smiles)
        standard_smiles=Chem.MolToSmiles(canonical_mol)
    except:
        standard_smiles=''
    Standard_Smiles.append(standard_smiles)
df['Standard_Smiles']=Standard_Smiles
df=df[df['Standard_Smiles']!='']
df.to_csv('All.csv',index=False)
df
```

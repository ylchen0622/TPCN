```python
import pandas as pd
from rdkit import Chem
```

### 1. Deglycosylation results

```python
df = pd.read_csv("deglycosylation_results.txt")
str1 = 'Nr;ID;originalMoleculeSMILES;deglycosylatedMoleculeSMILES;hadOrHasSugars;SugarMoietySMILES'
list1 = str1.split(';')
df1 = pd.DataFrame(columns=list1[1:])
df1
```

```python
def splite(x):
    p = x.split(";")[1:]
    return p
q1 = df["Nr;ID;originalMoleculeSMILES;deglycosylatedMoleculeSMILES;hadOrHasSugars;SugarMoietySMILES"].map(splite)
ID = []
originalMoleculeSMILES = []
deglycosylatedMoleculeSMILES = []
hadOrHasSugars = []
SugarMoietySMILES = []
for i in range(len(q1)):
    ID.append(q1[i][0])
    originalMoleculeSMILES.append(q1[i][1])
    deglycosylatedMoleculeSMILES.append(q1[i][2])
    hadOrHasSugars.append(q1[i][3])
    if len(q1[i])>=5:
        SugarMoietySMILES.append(q1[i][4:])
    else:
        SugarMoietySMILES.append([])
df1["ID"] = ID
df1["originalMoleculeSMILES"] = originalMoleculeSMILES
df1["deglycosylatedMoleculeSMILES"] = deglycosylatedMoleculeSMILES
df1["hadOrHasSugars"] = hadOrHasSugars
df1["SugarMoietySMILES"] = SugarMoietySMILES
df1
```

### 2. RDKit standardization

```python
originalMoleculeSMILES = []
deglycosylatedMoleculeSMILES = []
ID_list = []
for i in range(len(df1)):
    origi = df1['originalMoleculeSMILES'][i]
    degly = df1['deglycosylatedMoleculeSMILES'][i]
    try:
        ori_smi = Chem.MolToSmiles(Chem.MolFromSmiles(origi))
        originalMoleculeSMILES.append(ori_smi) 
        if degly == '[empty]':
            deglycosylatedMoleculeSMILES.append(degly)
        else:
            degly_smi = Chem.MolToSmiles(Chem.MolFromSmiles(degly))
            deglycosylatedMoleculeSMILES.append(degly_smi)
        ID_list.append(df1['ID'][i])
    except:
        pass
df1 = df1[df1.ID.isin(ID_list)]
df1['originalMoleculeSMILES'] = originalMoleculeSMILES
df1['deglycosylatedMoleculeSMILES'] = deglycosylatedMoleculeSMILES 
df1.to_csv("deglycosylation_results.csv",index=False)
df1
```

### 3. Add content information

```python
df_SRU=pd.read_csv('deglycosylation_results.csv')
df_All=pd.read_csv('../0 data preparation/All.csv',usecols=['ID','CAS Number','Type','Content(%)','Part','Family','Specie'])
df=df_All.merge(df_SRU,on='ID')
df.to_csv('All_SRU.csv',index=False)
df
```

### 4. Glycosylation ratio

### 4.1  Glycosylation ratio (All)

```python
df_HasSugar=df[df['hadOrHasSugars']==True].drop_duplicates(subset='ID')
df_NoSugar=df[df['hadOrHasSugars']==False].drop_duplicates(subset='ID')
df_Sugar_Percentage=len(df_HasSugar)/(len(df_HasSugar)+len(df_NoSugar))*100
Glycosides=[]
Glycosides.append((len(df_HasSugar),len(df_NoSugar),df_Sugar_Percentage))
Glycosides=pd.DataFrame(Glycosides,columns=['HasSugar','NoSugar','Percentage(%)'])
Glycosides['Percentage(%)']=round(Glycosides['Percentage(%)'],2)
Glycosides
```

### 4.2 Glycosylation ratio in each content range

```python
def Content_Glycosylation_Ratio(content_left,content_right):
    df_content=len(df[(df['Content(%)']>=content_left)&(df['Content(%)']<content_right)].drop_duplicates(subset='ID'))
    df_content_hassugar=len(df[(df['Content(%)']>=content_left)&(df['Content(%)']<content_right)&(df['hadOrHasSugars']==True)].drop_duplicates(subset='ID'))
    df_content_hassugar_percentages=round(df_content_hassugar/df_content*100,2)
    bins=str(content_left)+'~'+str(content_right)
    return bins,df_content_hassugar,df_content,df_content_hassugar_percentages
```

```python
content=[0.000001,0.00001,0.0001,0.001,0.01,0.1,1,10]
Data=[]
for i in range(len(content)):
    if i<=6:
        content_left=content[i]
        content_right=content[i+1]
        data=Content_Glycosylation_Ratio(content_left,content_right)
        Data.append(data)
    else:
        pass
Data=pd.DataFrame(Data,columns=['Bins', 'HasSugar', 'All', 'Percentages(%)'])
Data
```

### 4.3 Glycosylation ratio in low, medium and high content range

```python
content=[0.000001,0.0001,0.001,10]
Data1=[]
for i in range(len(content)):
    if i<=2:
        content_left=content[i]
        content_right=content[i+1]
        content_left=content[i]
        content_right=content[i+1]
        data=Content_Glycosylation_Ratio(content_left,content_right)
        Data1.append(data)
    else:
        pass
Data1=pd.DataFrame(Data1,columns=['Bins', 'HasSugar', 'All', 'Percentages(%)'])
Data1
```

### 4.4 Glycosylation ratio of each part

```python
def Part_Glycosylation_Ratio(location):
    df_location=len(df[df['Part']==location].drop_duplicates(subset='ID'))
    df_location_hassugar=len(df[(df['Part']==location)&(df['hadOrHasSugars']==True)].drop_duplicates(subset='ID'))
    df_location_hassugar_percentages=round(df_location_hassugar/df_location*100,2)
    return location,df_location_hassugar,df_location,df_location_hassugar_percentages
```

```python
Location=['Roots','Aerial parts','Whole plants','Leaves','Barks','Stems','Fruits','Seeds','Subterraneous stems','Flowers','Buds','Peels','Hooks','Woods','Husks','Seedling','Knots']
Part_GlyRatio=[]
for i in Location:
    part_glyratio=Part_Glycosylation_Ratio(i)
    Part_GlyRatio.append(part_glyratio)
Part_GlyRatio=pd.DataFrame(Part_GlyRatio,columns=['Part', 'HasSugar', 'All', 'Percentages(%)'])
Part_GlyRatio.sort_values(by=['Percentages(%)'],ascending=False,inplace=True)
Part_GlyRatio.reset_index(drop=True,inplace=True)
Part_GlyRatio
```
